# Text Preprocessing with NLTK and spaCy

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import spacy

# Download necessary NLTK datasets
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Load spaCy model
get_ipython().system('python -m spacy download en_core_web_sm')

# Sample text
text = "NLTK and spaCy are powerful NLP libraries for text processing."

# NLTK Processing
tokens_nltk = word_tokenize(text)
stop_words = set(stopwords.words('english'))
filtered_tokens_nltk = [word for word in tokens_nltk if word.lower() not in stop_words]
lemmatizer = WordNetLemmatizer()
lemmatized_tokens_nltk = [lemmatizer.lemmatize(word) for word in filtered_tokens_nltk]

# spaCy Processing
nlp = spacy.load("en_core_web_sm")
doc = nlp(text)
tokens_spacy = [token.text for token in doc]
filtered_tokens_spacy = [token.text for token in doc if not token.is_stop]
lemmatized_tokens_spacy = [token.lemma_ for token in doc if not token.is_stop]

# Print Results
print("NLTK Tokenization:", tokens_nltk)
print("NLTK Stopword Removal:", filtered_tokens_nltk)
print("NLTK Lemmatization:", lemmatized_tokens_nltk)
print("\nspaCy Tokenization:", tokens_spacy)
print("spaCy Stopword Removal:", filtered_tokens_spacy)
print("spaCy Lemmatization:", lemmatized_tokens_spacy)
```

# Named Entity Recognition (NER) using spaCy

```python
import spacy
from spacy import displacy

# Sentence about Kazakhstan
text = "Kazakhstan, the largest landlocked country in the world, gained independence from the Soviet Union in 1991. Its capital, Astana, was formerly known as Nur-Sultan."

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Process text
doc = nlp(text)

# Print named entities
for ent in doc.ents:
    print(ent.text, ent.label_)

# Visualizing named entities
displacy.render(doc, style="ent", jupyter=True)
```

# Text Vectorization using Transformers

```python
from transformers import AutoTokenizer, AutoModel
import torch

# Load tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
model = AutoModel.from_pretrained("bert-base-uncased")

# Unique sentence about Kazakhstan
sentence = "Kazakhstan is known for its vast steppes, rich cultural heritage, and the Baikonur Cosmodrome, the world's first and largest space launch facility."

# Tokenize input sentence
inputs = tokenizer(sentence, return_tensors="pt", padding=True, truncation=True)

# Extract embeddings
with torch.no_grad():
    outputs = model(**inputs)

# Extract word embeddings
embeddings = outputs[0]  # First element contains embeddings
print("Embedding Shape:", embeddings.shape)  # (batch_size, seq_length, hidden_dim)
```

# Sentiment Analysis with Transformers

```python
from transformers import pipeline

# Load sentiment analysis pipeline
sentiment_analyzer = pipeline("sentiment-analysis")

# New interesting sentences
sentences = [
    "Kazakhstan has made remarkable progress in technology and space research.",
    "The Baikonur Cosmodrome has witnessed historic space launches, but some missions have ended in failure.",
    "The landscapes of Kazakhstan are breathtaking, attracting tourists from around the world.",
    "The economic crisis has led to increased unemployment and uncertainty.",
    "The recent scientific breakthrough in Kazakhstan's renewable energy sector is truly inspiring."
]

# Analyze sentiment
results = sentiment_analyzer(sentences)

# Display results
for sent, result in zip(sentences, results):
    print(f"Sentence: {sent}")
    print(f"Sentiment: {result['label']}, Score: {result['score']:.4f}\n")
```
